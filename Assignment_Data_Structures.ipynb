**1.  Discuss String Slicing and provide examples.**

String slicing in Python is a technique used to extract a portion (substring) of a string by specifying a range of indices.

**The syntax for slicing is:-**

    string[start:stop:step]

**start:-** The starting index of the slice (inclusive). If omitted, it defaults to 0.

**stop:-** The ending index of the slice (exclusive). If omitted, it defaults to the length of the string.

**step:-** The step size or interval between each character in the slice. If omitted, it defaults to 1.

**Examples of String Slicing**

**a. Basic Slicing:-**

    text = "Hello, World!"
    print(text[0:5])
    # Output: 'Hello'

This extracts the substring from index 0 to 4 (since the stop index is exclusive).

**b. Omitting Start or Stop:-**

    text = "Hello, World!"
    print(text[:5])   # Output: 'Hello' (equivalent to text[0:5])
    print(text[7:])   # Output: 'World!' (from index 7 to the end)

Omitting the start or stop will take the substring from the beginning or until the end of the string, respectively.

**c. Using Step:-**

    text = "Hello, World!"
    print(text[::2])  
    # Output: 'Hlo ol!' (every second character)

The step size of 2 skips every other character in the string.

**d. Negative Indices:-**

    text = "Hello, World!"
    print(text[-6:-1])  # Output: 'World' (from index -6 to -2)

Negative indices count from the end of the string, where -1 is the last character.

**e. Reversing a String:-**


    text = "Hello, World!"
    print(text[::-1])  # Output: '!dlroW ,olleH'

Using a negative step of -1 reverses the string.

**Practical Uses of String Slicing**

**Extracting parts of a URL:-**

    url = "https://www.example.com"
    domain = url[8:-4]  # Output: 'www.example'

**Checking for Palindromes:-**

    word = "radar"
    if word == word[::-1]:
    print("It's a palindrome!")

String slicing is a powerful tool that can be used for a variety of purposes, from basic substring extraction to more complex operations like reversing or selecting specific patterns within strings.


**2. Explain the key features of lists in python**

In Python, lists are one of the most versatile and commonly used data structures. They allow you to store collections of items in a single variable and provide a wide range of features that make them highly useful.

Below are some key features of lists in Python:

**a. Ordered Collection**

Lists maintain order: The order in which elements are added to a list is preserved. When you iterate over a list, elements are returned in the order they were added.

**Example:-**

    fruits = ["apple", "banana", "cherry"]
    print(fruits)  # Output: ['apple', 'banana', 'cherry']
**b. Mutable**

Lists are mutable: This means you can change, add, or remove elements after the list has been created.

**Example:-**

    fruits = ["apple", "banana", "cherry"]
    fruits[1] = "blueberry"
    print(fruits)  # Output: ['apple', 'blueberry', 'cherry']
**c. Heterogeneous Elements**

Lists can contain different types of elements: A single list can hold integers, strings, floats, and even other lists.

**Example:-**

    mixed_list = [1, "Hello", 3.14, [1, 2, 3]]
    print(mixed_list)  # Output: [1, 'Hello', 3.14, [1, 2, 3]]
**d. Dynamic Size**

Lists can grow and shrink: Unlike arrays in some other languages, lists in Python can dynamically change size, meaning you can add or remove elements as needed.

**Example:-**

    fruits = ["apple", "banana"]
    fruits.append("cherry")
    print(fruits)  # Output: ['apple', 'banana', 'cherry']

    fruits.pop(1)
    print(fruits)  # Output: ['apple', 'cherry']
**e. Indexing and Slicing**

Lists support indexing: You can access individual elements of a list using their index. Python uses zero-based indexing.
Lists also support slicing: You can extract parts of the list using slicing.

**Example:-**

    fruits = ["apple", "banana", "cherry"]
    print(fruits[1])     # Output: 'banana'
    print(fruits[0:2])   # Output: ['apple', 'banana']

**f. List Comprehensions**

Compact way to create lists: List comprehensions provide a concise way to create lists by embedding loops and conditions inside square brackets.

**Example:-**

    squares = [x**2 for x in range(5)]
    print(squares)  # Output: [0, 1, 4, 9, 16]
**g. Common List Methods**

    append():- Adds an item to the end of the list.
    extend():- Adds all elements of an iterable (e.g., another list) to the end of the list.
    insert():- Inserts an item at a given position.
    remove():- Removes the first occurrence of a specified item.
    pop():- Removes and returns the item at the given index (default is the last item).
    index():- Returns the index of the first occurrence of a specified item.
    count():- Returns the number of occurrences of a specified item.
    sort():- Sorts the list in place.
    reverse():- Reverses the elements of the list in place.
**Example of some methods:-**

    numbers = [1, 2, 3, 4, 3]
    numbers.append(5)
    print(numbers)  # Output: [1, 2, 3, 4, 3, 5]

    numbers.remove(3)
    print(numbers)  # Output: [1, 2, 4, 3, 5]

    numbers.sort()
    print(numbers)  # Output: [1, 2, 3, 4, 5]
**h. Nesting**

Lists can contain other lists: You can create nested lists, which are lists within lists. This allows for the creation of multi-dimensional data structures, like matrices.

**Example:-**

    matrix = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
    print(matrix[1][2])  # Output: 6
**i. Iteration**

You can easily loop through lists: Using loops like for, you can iterate over the elements of a list.

**Example:-**

    fruits = ["apple", "banana", "cherry"]
    for fruit in fruits:
    print(fruit)
    # Output:
    # apple
    # banana
    # cherry
**j. Copying Lists**

Copying lists requires care: If you assign one list to another, both refer to the same list in memory. Use slicing or the copy() method to create a copy.

**Example:-**

    original = [1, 2, 3]
    copy = original[:]
    copy.append(4)
    print(original)  # Output: [1, 2, 3]
    print(copy)      # Output: [1, 2, 3, 4]
These features make lists a powerful and flexible tool in Python for storing and manipulating data.

**3. Describe how to access, modify and delete elements in a list with examples**

Accessing, Modifying, and Deleting Elements in a List

**a. Accessing Elements**

You can access elements in a list using indexing. Python lists are zero-indexed, meaning the first element has an index of 0.

**Example of Accessing Elements:-**

    fruits = ["apple", "banana", "cherry", "date"]

    # Access the first element
    first_fruit = fruits[0]
    print("First fruit:", first_fruit)  
    # Output: 'apple'

    # Access the last element
    last_fruit = fruits[-1]
    print("Last fruit:", last_fruit)  
    # Output: 'date'

    # Access a range of elements (slicing)
    middle_fruits = fruits[1:3]
    print("Middle fruits:", middle_fruits)  # Output: ['banana', 'cherry']

**b. Modifying Elements**

You can modify elements in a list by assigning a new value to a specific index.

**Example of Modifying Elements:-**

    fruits = ["apple", "banana", "cherry", "date"]

    # Change the second element
    fruits[1] = "blueberry"
    print("Modified list:", fruits)  
    # Output: ['apple', 'blueberry', 'cherry', 'date']

    # Modify a range of elements
    fruits[1:3] = ["blackberry", "cranberry"]
    print("Modified range:", fruits)  
    # Output: ['apple', 'blackberry', 'cranberry', 'date']
**c. eleting Elements**

You can delete elements from a list using the del statement, pop() method, or remove() method.

**Using del Statement:-**

del removes an element by index or deletes a slice of the list.

**Example of Deleting Elements using del:-**

    fruits = ["apple", "banana", "cherry", "date"]

    # Delete the second element
    del fruits[1]
    print("After del:", fruits)  
    # Output: ['apple', 'cherry', 'date']

    # Delete a range of elements
    del fruits[1:3]
    print("After deleting a range:", fruits)  
    # Output: ['apple']
**Using pop() Method:-**

pop() removes and returns the element at the given index. If no index is provided, it removes the last element.

**Example of Deleting Elements using pop:-**

    fruits = ["apple", "banana", "cherry", "date"]

    # Remove and return the last element
    last_fruit = fruits.pop()
    print("Popped fruit:", last_fruit)  
    # Output: 'date'
    print("After pop:", fruits)  
    # Output: ['apple', 'banana', 'cherry']

    # Remove and return the first element
    first_fruit = fruits.pop(0)
    print("Popped first fruit:", first_fruit)  
    # Output: 'apple'
    print("After pop at index 0:", fruits)  
    # Output: ['banana', 'cherry']

**Using remove() Method:-**

remove() removes the first occurrence of the specified element. If the element is not found, it raises a ValueError.

**Example of Deleting Elements using remove:-**

    fruits = ["apple", "banana", "cherry", "banana"]

    # Remove the first occurrence of "banana"
    fruits.remove("banana")
    print("After remove:", fruits)  
    # Output: ['apple', 'cherry', 'banana']

    # Remove the second "banana"
    fruits.remove("banana")
    print("After removing second 'banana':", fruits)  
    # Output: ['apple', 'cherry']

These methods give you control over the contents of your lists, allowing you to manage the data stored within them effectively.

**5. Describe the key features of sets and provide examples of their use.**

Sets in Python are a unique data structure that stores unordered collections of unique elements. They are commonly used when you need to eliminate duplicate values or perform mathematical set operations like unions and intersections. Here's an overview of the key features of sets, along with examples of their use.

**Key Features of Sets in Python**

**a. Unordered Collection**

**Sets are unordered:-**

The elements in a set do not have a specific order, and you cannot access elements by index like you can with lists or tuples.

**Example:-**

    my_set = {3, 1, 4, 2}
    print(my_set)  
    # Output could be {1, 2, 3, 4} or any other permutation
**b.Unique Elements**

Sets automatically eliminate duplicates: If you try to add duplicate elements to a set, it will store only one instance of each element.

**Example:-**

    my_set = {1, 2, 2, 3, 4, 4}
    print(my_set)  
    # Output: {1, 2, 3, 4}
**c.Mutable**

Sets are mutable: You can add or remove elements from a set after it is created. However, the elements themselves must be immutable (e.g., numbers, strings, tuples).

**Example:-**

    my_set = {1, 2, 3}
    my_set.add(4)
    print(my_set)  
    # Output: {1, 2, 3, 4}

    my_set.remove(2)
    print(my_set)  
    # Output: {1, 3, 4}
**d.No Duplicates**

Sets inherently do not allow duplicates: This feature is particularly useful when you need to remove duplicate items from a list or any other iterable.

**Example:-**

    my_list = [1, 2, 2, 3, 4, 4]
    my_set = set(my_list)
    print(my_set)  # Output: {1, 2, 3, 4}
**e.Efficient Membership Testing**
Sets provide fast membership testing: Checking if an item is in a set is much faster compared to lists, especially for large datasets.

**Example:-**

    my_set = {1, 2, 3, 4, 5}
    print(3 in my_set)  
    # Output: True
    print(10 in my_set)  
    # Output: False
**f.Set Operations**

Sets support mathematical operations: You can perform union, intersection, difference, and symmetric difference operations, which are common in mathematical set theory.

**Union (|):-** Combines all elements from two sets, without duplicates.

**Intersection (&):-** Returns only the elements that are common to both sets.

**Difference (-):-** Returns the elements that are in the first set but not in the second.

**Symmetric Difference (^):-** Returns the elements that are in either set, but not in both.

**Example:-**

    set_a = {1, 2, 3, 4}
    set_b = {3, 4, 5, 6}

    print("Union:", set_a | set_b)  # Output: {1, 2, 3, 4, 5, 6}
    print("Intersection:", set_a & set_b)  # Output: {3, 4}
    print("Difference (A - B):", set_a - set_b)  # Output: {1, 2}
    print("Symmetric Difference:", set_a ^ set_b)  # Output: {1, 2, 5, 6}
**g. Immutability with frozenset**

frozenset is an immutable version of set: Once created, elements cannot be added or removed from a frozenset.

**Example:-**

    frozen_set = frozenset([1, 2, 3, 4])
    print(frozen_set)  # Output: frozenset({1, 2, 3, 4})

    # frozen_set.add(5)  # This would raise an AttributeError

When you have a list with duplicate items and need a list of unique items:

**8.Discuss the importance of dictionary keys being immutable and provide examples.**

In Python, dictionary keys must be immutable. This immutability requirement is crucial for the internal mechanics of how dictionaries operate, particularly for efficient key lookups, hash-based storage, and maintaining the integrity of the data structure. Let's discuss why dictionary keys must be immutable and provide examples to illustrate this concept.

**Why Dictionary Keys Must Be Immutable**

**a. Hashing Requirement**

**Explanation:-**
Python dictionaries use a technique called hashing to store and retrieve values efficiently. A hash function takes an object (in this case, a dictionary key) and returns a fixed-size integer (the hash value) that uniquely identifies the object. For the hashing process to work consistently, the object used as a key must not change after it is hashed.

**Example:-**
If a mutable object like a list were used as a key and it changed, the hash value would change, and the dictionary wouldn't be able to find the key, leading to errors or incorrect data retrieval.

    # Attempting to use a list as a key (This will raise a TypeError)
    my_dict = {}
    my_list = [1, 2, 3]
    # my_dict[my_list] = "value"  # Uncommenting this line will raise an error
**b.Consistency and Integrity**

**Explanation:-** The immutability of keys ensures that the dictionary's data structure remains consistent. If a key could change after being added to the dictionary, it would undermine the integrity of the dictionary, as the key might no longer map to the correct value.

**Example:-** Consider a scenario where you store an object in a dictionary, and then later, the object (if it were mutable) is modified. The dictionary would no longer be able to locate the correct value because the key has effectively changed.

    # Example where immutability ensures consistency
    my_dict = {}
    my_tuple = (1, 2, 3)
    my_dict[my_tuple] = "Tuple as key"

    # If my_tuple were mutable and changed to (4, 5, 6), the dictionary lookup would fail
    # But since tuples are immutable, the key remains valid
    print(my_dict[my_tuple])  # Output: "Tuple as key"
**c. Performance Optimization**
**Explanation:-** Immutable keys allow Python dictionaries to perform lookups, insertions, and deletions more efficiently. The immutability guarantees that the hash value calculated at the time of insertion remains valid throughout the key's lifetime in the dictionary.

**Example:-** Immutable types like strings and numbers can be quickly hashed and compared, making operations on dictionaries fast.

    # Using an integer as a key (immutable)
    my_dict = {1: "one", 2: "two"}
    print(my_dict[1])  # Output: "one"

    # Using a string as a key (immutable)
    my_dict = {"name": "Alice", "age": 30}
    print(my_dict["name"])  # Output: "Alice"

**8. Discuss the importance of dictionary keys being immutable and provide examples.**

**Importance of Dictionary Keys Being Immutable**

**a. Hashing Requirement**

**Explanation:-** Python dictionaries use a technique called hashing to store and retrieve values. A hash function converts an object (in this case, a dictionary key) into a fixed-size integer (the hash value) that uniquely identifies the object. This hash value is used to quickly locate the corresponding value in the dictionary.

For hashing to work correctly, the key must remain consistent throughout its lifetime in the dictionary. If the key could change (as mutable objects can), the hash value would also change, causing the dictionary to lose track of the key-value pair, leading to potential data corruption or inability to find the key.

**Example:-**

    # Example where immutability is required
    my_dict = {}
    my_tuple = (1, 2, 3)
    my_dict[my_tuple] = "Tuple as key"

    # If tuples were mutable and my_tuple changed, the dictionary would fail to locate the key
    print(my_dict[my_tuple])  # Output: "Tuple as key"
**b. Consistency and Integrity**

**Explanation:-** The immutability of keys ensures that once a key is inserted into a dictionary, it will consistently map to the same value. If dictionary keys were mutable, changing the key after its been added to the dictionary would disrupt the integrity of the data structure, as the original key would no longer map to the correct value.

This immutability is crucial for the internal workings of the dictionary, allowing it to maintain a stable and predictable association between keys and values.

**Example:-**

    # Using a string (immutable) as a key ensures the key-value pair remains consistent
    my_dict = {"name": "Alice"}
    print(my_dict["name"])  # Output: "Alice"

    # If we try to use a mutable object like a list as a key, it would cause issues
    # my_list = [1, 2, 3]
    # my_dict[my_list] = "This would raise a TypeError"
**c. Efficiency and Performance**

**Explanation:-** The immutability of dictionary keys allows for efficient hashing and lookup operations. Since the keys hash value is calculated once (when the key is first inserted), this value can be reused whenever the key is accessed in the dictionary. If keys were mutable, the hash would need to be recalculated every time the key changed, significantly reducing performance.

Immutable keys like integers, strings, and tuples allow Python to optimize these operations, making dictionaries one of the most efficient data structures for quick lookups.

**Example:-**

    # Using an integer as a key (efficient and immutable)
    my_dict = {1: "one", 2: "two"}
    print(my_dict[1])  # Output: "one"

    # Using a string as a key (efficient and immutable)
    my_dict = {"name": "Alice", "age": 30}
    print(my_dict["name"])  # Output: "Alice"
